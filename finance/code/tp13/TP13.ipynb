```python
import math

import matplotlib.pyplot as plt
```

```python
def compute_futures_price(asset_price, remaining_time, risk_free_rate=0.1):
    return asset_price * math.exp(risk_free_rate * remaining_time)
```

```python
months = [0, 2, 4, 6, 8, 10, 12]
prices_1 = [100, 105, 104, 108, 113, 109, 112]
prices_2 = [100, 99, 101, 98, 97, 99, 97]
futures_prices_1 = [compute_futures_price(x, (12 - m) / 12) for m, x in zip(months, prices_1)]
futures_prices_2 = [compute_futures_price(x, (12 - m) / 12) for m, x in zip(months, prices_2)]
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(months, prices_1, 'rs-', label='Price (scenario 1)')
ax.plot(months, futures_prices_1, 'rs--', label='Futures (scenario 1)')
ax.plot(months, prices_2, 'gs-', label='Price (scenario 2)')
ax.plot(months, futures_prices_2, 'gs--', label='Futures (scenario 2)')
ax.set_xlabel('Time (months)')
ax.legend()
plt.show()
```
